```python
# Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

def frecuencia_letras(texto):
    resultado = {}
    for letra in texto:
        if letra != " ":
            if letra in resultado:
                resultado[letra] = resultado[letra] + 1
            else:
                resultado[letra] = 1
    return resultado

texto = input("Escribe un texto: ")
frecuencia = frecuencia_letras(texto)
print(frecuencia)
```

```python
# 13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()

def letras_mayus_minus(cadena):
    unicos = []
    for caracteres in cadena:
        if caracteres not in unicos:
            unicos.append(caracteres)

    solo_letras = list(filter(str.isalpha, unicos))

    return list(map(lambda c: (c.upper(), c.lower()), solo_letras))

letras_mayus_minus("Hola, Hola, 123!")
```

```python
# 16. Escribe una función que tome una cadena de texto y un número entero n como parámetros y devuelva una lista de todas las palabras que sean más largas que n. Usa la función filter().

def cadena_texto(texto,n):
    palabras = texto.split()

    def longitud(p):
        return len(p) > n
    
    return list(filter(longitud,palabras))

print(cadena_texto("Esto es una prueba de un ejercicio en python", 4))
print(cadena_texto("Esto es una prueba de un ejercicio en python", 2))
```

```python
# 22. Dada una lista numérica, obtén el producto total de los valores de dicha lista.Usa la función reduce() .

from functools import reduce

def producto_total(numeros):
    def multiplicar(valores, n):
        return valores * n
    return reduce(multiplicar, numeros, 1)

print(producto_total([2,4]))
print(producto_total([3,5,6,9,15]))
print(producto_total([]))
```

```python
#28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

def primer_duplicado(lista):
    repetido = set()
    for elemento in lista:
        if elemento in repetido:
            return elemento
        repetido.add(elemento)

print(primer_duplicado([1,2,3,4,5,4,6,2]))
```

```python
#30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

def anagramas(palabra1, palabra2):
    p1 = palabra1.replace(" ","").lower()
    p2 = palabra2.replace(" ","").lower()

    return sorted(p1) == sorted(p2)

print(anagramas("Roma", "Amor")) 
```
