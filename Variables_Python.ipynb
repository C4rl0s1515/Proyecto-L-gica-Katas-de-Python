```python
# 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

def frecuencia_letras(texto):
    texto = texto.lower()
    resultado = {}
    for letra in texto:
        if letra != " ":
            if letra in resultado:
                resultado[letra] = resultado[letra] + 1
            else:
                resultado[letra] = 1
    return resultado

texto = input("Escribe un texto: ")
frecuencia = frecuencia_letras(texto)
print(frecuencia)
```

```python
# 13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()

def letras_mayus_minus(cadena):
   
    letras = set(c.lower() for c in cadena if c.isalpha())
    
    return list(map(lambda c: (c.upper(), c.lower()), letras))

resultado = letras_mayus_minus("Hola, Hola, Sol, noches, 123!, prueba")
print(resultado)

# Se ha simplificado la función utilizando un set para gestionar los duplicados de forma más eficiente, ya que los conjuntos en Python eliminan automáticamente los valores repetidos, evitando así tener que usar listas auxiliares y bucles manuales.
```

```python
#28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

def primer_duplicado(lista):
    repetido = set()
    for elemento in lista:
        if elemento in repetido:
            return elemento
        repetido.add(elemento)

resultado = primer_duplicado([1,2,3,4,5,4,6,2])
resultado2 = primer_duplicado(["a","b","x","z","b","a"])

print(resultado)
print(resultado2)
```

```python
#30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

def anagramas(palabra1, palabra2):
    p1 = palabra1.replace(" ","").lower()
    p2 = palabra2.replace(" ","").lower()

    return sorted(p1) == sorted(p2)

resultado = anagramas("Roma ana", "ana Amor")
resultado2 = anagramas("Casa", "asac")

print(resultado)
print(resultado2)
```

```python
# Codigo a seguir

class Arbol:
    def __init__(self, tronco, ramas):
        self.tronco = tronco
        self.ramas = ramas

    def crecer_tronco(self):
        self.tronco += 1
    
    def nueva_rama(self):
        self.ramas.append(1)

    def crecer_ramas(self):
        for r in range(len(self.ramas)):
            self.ramas[r] +=1

    def quitar_rama(self, posicion):
        if 0 <= posicion < len(self.ramas):
           del self.ramas[posicion]
        else:
            return IndexError("Posición de rama no válida")
            
    def info_arbol(self):
        return (
            f"longitud_tronco: {self.tronco} ," 
            f"numero_ramas: {len(self.ramas)} ,"
            f"longitud_ramas: {self.ramas}" 
            )
    

# 1. Crear un árbol.
a = Arbol(1, [])
print(a.info_arbol())

# 2. Hacer crecer el tronco del árbol una unidad.
a = Arbol(1, [])
a.crecer_tronco()
print(a.info_arbol())
    
# 3. Añadir una nueva rama al árbol.
a = Arbol(1, [])
a.crecer_tronco()
a.nueva_rama()
print(a.info_arbol())

# 4. Hacer crecer todas las ramas del árbol una unidad.
a = Arbol(1, [])
a.crecer_tronco()
a.nueva_rama()
a.crecer_ramas()
print(a.info_arbol())

# 5. Añadir dos nuevas ramas al árbol.
a = Arbol(1, [])
a.crecer_tronco()
a.nueva_rama()
a.crecer_ramas()
a.nueva_rama()
a.nueva_rama()
print(a.info_arbol())

# 6. Retirar la rama en la posición 2.
a = Arbol(1, [])
a.crecer_tronco()
a.nueva_rama()
a.crecer_ramas()
a.nueva_rama()
a.nueva_rama()
a.quitar_rama(2)
print(a.info_arbol())

# 7. Obtener información sobre el árbol.
    
print(a.info_arbol())
```

```python
# 37.  Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras, reemplazar_palabras, eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto.

def contar_palabras(texto):
    palabras = texto.split()
    cuenta = {}

    for palabra in palabras:
        if palabra in cuenta:
            cuenta[palabra] += 1
        else:
            cuenta[palabra] = 1
    return cuenta

def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    escrito = texto.split()
    remplazo = [palabra_nueva if t == palabra_original else t for t in escrito]
    return " ".join(remplazo)

def eliminar_palabra(texto, palabra_eliminar):
    palabras = texto.split()
    suprimir = [p for p in palabras if p != palabra_eliminar]
    return " ".join(suprimir)

def procesar_texto(texto, eleccion, *args):
    if eleccion == "contar":
        return contar_palabras(texto)
    
    elif eleccion == "remplazar":
        palabra_original, palabra_nueva = args
        return reemplazar_palabras(texto, palabra_original, palabra_nueva)
    
    elif eleccion == "eliminar":
         palabra_eliminar = args[0]
         return eliminar_palabra(texto, palabra_eliminar)
    
    else: 
        return "Error, opción no valida"
    

t = "hola buen día esto es la primera prueba del día prueba día hola"

print(procesar_texto( t, "contar"))

print(procesar_texto( t , "remplazar", "día", "noche"))

print(procesar_texto( t , "eliminar", "día"))
    
```

```python
# 40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo") datos (una tupla con los datos necesarios para calcular el área de la figura).

from math import pi

def area (figura, datos):
    if figura == "rectangulo":
        base, altura = datos
        return round(base * altura , 2)
    
    elif figura == "triangulo":
        base, altura = datos
        return round((base * altura)/2 , 2)
    
    elif figura == "circulo":
        (radio,) = datos
        return round(pi * (radio**2), 2)

    else:
         return "Error, figura no valida"
    
print(area ("rectangulo", (3.5555,5.89471)))
print(area ("triangulo", (6.55,5.89)))
print(area ( "circulo", (5,)))
```
